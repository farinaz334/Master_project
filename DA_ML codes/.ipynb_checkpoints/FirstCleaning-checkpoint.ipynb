# Functions

```python
def df_info(df):
    print(df.head(10))
    print("\n--------------------------------------------------\n")
    print(df.info())
```

```python
def df_type(df ,numerical_object_features ):
    #print(df.dtypes)

    for feature in numerical_object_features:
        df[feature] = pd.to_numeric(df[feature], errors='coerce')

    print(df.dtypes)
```

```python
#Note: This function is designed to help make decisions in handling missing data


def df_check_null_values(df ,selected_features):
    
    for col in selected_features:
        df[col] = pd.to_numeric(df[col])
        
    print("SUM of missing values:")
    print(df.isnull().sum())
    print("\n--------------------------------------------------\n")
    total_rows_df = df.shape[0]
    missing_percentage_df = (df.isnull().sum() / total_rows_df) * 100
    missing_percentage_df = round(missing_percentage_df, 2)
    print("Percentage of missing values:")
    print(missing_percentage_df)
    print("\n--------------------------------------------------\n")
    # Creating a selected features list to have a look on their distributions to  choose an appropriate imputation method ...
    selected_stats = df[selected_features].dropna().describe()
    print(selected_stats)
    print("\n--------------------------------------------------\n")
```

```python
#def boxplot(df ,selected_features):
#    plt.figure(figsize=(10, 12))
#    sns.boxplot(data=df[selected_features])
#    plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
#    plt.title('Boxplot for Each Column')
#    plt.xlabel('Columns')
#    plt.ylabel('Values')
#    plt.show()




def boxplot(df, selected_features, features_per_plot=5):
    num_features = len(selected_features)
    num_plots = (num_features + features_per_plot - 1) // features_per_plot  # Calculate the number of plots needed

    for i in range(num_plots):
        plt.figure(figsize=(10, 12))
        subset_features = selected_features[i*features_per_plot:(i+1)*features_per_plot]
        sns.boxplot(data=df[subset_features])
        plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
        plt.title(f'Boxplot for Features {i*features_per_plot+1} to {(i+1)*features_per_plot}')
        plt.xlabel('Columns')
        plt.ylabel('Values')
        plt.show()
```

```python
#def histogram(df ,selected_features):
#    plt.figure(figsize=(12, 8))
#    row =  math.ceil(len(selected_features)/3)
#    for i, feature in enumerate(selected_features):
#        plt.subplot(row, 3, i+1)
#        sns.histplot(df[feature], kde=True)
#        plt.title(feature)

#    plt.tight_layout()
#    plt.show()




def histogram(df, selected_features, save_file="histograms.png"):
    num_features = len(selected_features)
    num_columns = 3
    num_rows = math.ceil(num_features / num_columns)

    plt.figure(figsize=(12, num_rows * 4))  # Adjust height based on the number of rows

    for i, feature in enumerate(selected_features):
        plt.subplot(num_rows, num_columns, i + 1)
        sns.histplot(df[feature], kde=True)
        plt.title(feature)
        plt.xlabel('')
        plt.ylabel('')

    plt.tight_layout()

    # Save the plot to a file
    plt.savefig(save_file)
    print(f"Histogram saved as {save_file}")

    # Display the plot interactively
    plt.show()



```

```python
def imputation(df, numerical_features, method):
    method = method.lower()
    
    # Impute missing values for numerical features
    for feature in numerical_features:
        if method == "median":
            df[feature].fillna(df[feature].median(), inplace=True)
        elif method == "mode":
            modes = df[feature].mode()  # Calculate mode for the feature
            mode1 = modes[0]  # First mode
            mode2 = modes[1] if len(modes) > 1 else modes[0]  # Second mode if available
            
            # Find indices of missing values
            null_indices = df[df[feature].isnull()].index.tolist()  # Convert Index to list
            
            # Calculate half size
            half_size = len(null_indices) // 2
            
            # Randomly shuffle the indices
            np.random.shuffle(null_indices)
            
            # Replace missing values with mode1 and mode2
            df.loc[null_indices[:half_size], feature] = mode1
            df.loc[null_indices[half_size:], feature] = mode2
        elif method == "mean":
            df[feature].fillna(df[feature].mean(), inplace=True)
        else:
            print(f"Unknown method '{method}' for imputation.")
    
    # Print dataframe information after imputation
    print(df.info())
```

```python
def duplicate_rows(df, save_csv='no'):
    duplicate_rows_count = df.duplicated().sum()
    print("Duplicate rows:", duplicate_rows_count)
    

    
    if save_csv.lower() == 'yes':
        df[df.duplicated()].to_csv('duplicate_rows.csv', index=False)
        print("Duplicate rows saved to 'duplicate_rows.csv'")
    if duplicate_rows_count > 0:
        
       df = df.drop_duplicates()
    return df
```

```python
def drop_col(df, targeted_cols):
    print("Targeted columns to drop:", targeted_cols)
    df.drop(columns=targeted_cols, inplace=True)
    print("Remaining columns:", df.columns)

```
